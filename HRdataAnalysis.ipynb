```python
# Import dependecies
import pandas as pd
```

```python
# Load csv file
df=pd.read_csv("/content/HR Data.csv")
df.head()
```

```python
# Display the initial structure of the dataset
print("Initial structure:")
print(df.info())
```

```python
# 1. Remove unnecessary columns
unnecessary_columns = ["DailyRate", "Education","EmployeeCount","MonthlyRate","RelationshipSatisfaction"]  # Specify columns to remove
df.drop(columns=unnecessary_columns, inplace=True)
```

```python
# 1. Remove unnecessary columns
unnecessary_columns = ["Attrition","HourlyRate"]  # Specify columns to remove
df.drop(columns=unnecessary_columns, inplace=True)
```

```python
df
```

```python
# 2. Give the columns new names
new_column_names = {"DistanceFromHome": "DistanceFromHome(km)","EmployeeNumber":"EmployeeID","MonthlyIncome":"MonthlyIncome(Rs)",}  # Specify new names
df.rename(columns=new_column_names, inplace=True)
```

```python
df
```

```python
df.duplicated()
```

```python
# 3. Eliminate redundant entries
df.drop_duplicates(inplace=True)
```

```python
# 4. Sanitize specific columns

df['Over18'] = df['Over18'].str.replace('Y', 'Yes')
df['Over18'] = df['Over18'].str.replace('N', 'No')
```

```python
df.isna().sum()
```

```python
# 5. Eliminate NaN values
df.dropna(inplace=True)
```

```python
# Display the final structure of the dataset
print("\nFinal structure:")
print(df.info())
```

```python
df.to_csv('modified_HR_Data.csv')
```

```python
# Selecting columns of interest
selected_columns = ['Age', 'BusinessTravel', 'Department', 'DistanceFromHome(km)', 'EducationField',
                    'EnvironmentSatisfaction', 'Gender', 'JobInvolvement', 'JobLevel', 'JobRole',
                    'JobSatisfaction', 'MaritalStatus', 'MonthlyIncome(Rs)', 'NumCompaniesWorked',
                    'OverTime', 'PercentSalaryHike', 'PerformanceRating', 'StockOptionLevel',
                    'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany',
                    'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']

# Creating a DataFrame with selected columns
selected_data = df[selected_columns]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Visualizing Department distribution
plt.figure(figsize=(10, 6))
sns.countplot(y='Department', data=selected_data, palette='muted')
plt.title('Department Distribution')
plt.xlabel('Count')
plt.ylabel('Department')
plt.grid(True)
plt.show()
```

```python
# Plotting years of experience vs salary
plt.figure(figsize=(8, 6))
plt.scatter(df['Department'], df['MonthlyIncome(Rs)'], color='orange')
plt.title('Department vs MonthlyIncome')
plt.xlabel('Department')
plt.ylabel('MonthlyIncome(Rs)')
plt.grid(True)
plt.show()
```

```python
# Visualizing Monthly Income distribution by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='MonthlyIncome(Rs)', data=selected_data, palette='pastel')
plt.title('Monthly Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Monthly Income (Rs)')
plt.grid(True)
plt.show()
```

```python
# Plotting performance rating distribution
plt.figure(figsize=(8, 6))
plt.bar(df['EmployeeID'], df['PerformanceRating'], color='green')
plt.title('Performance Rating Distribution')
plt.xlabel('Employee ID')
plt.ylabel('Performance Rating')
plt.grid(axis='y')
plt.show()
```

```python
# Visualizing Job Role distribution
plt.figure(figsize=(12, 8))
sns.countplot(y='JobRole', data=selected_data, palette='bright')
plt.title('Job Role Distribution')
plt.xlabel('Count')
plt.ylabel('Job Role')
plt.grid(True)
plt.show()
```
